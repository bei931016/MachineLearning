```python
from sklearn import tree
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

# 讀入鳶尾花資料
iris = load_iris()
iris_data = iris.data
iris_label = iris.target

# 切分訓練與測試資料
train_data, test_data, train_label, test_label = train_test_split(iris_data, iris_label, test_size = 0.3)

# 建立分類器
clf = tree.DecisionTreeClassifier()
iris_clf = clf.fit(train_data, train_label)

print("訓練資料正確率:",clf.score(train_data,train_label)) # 訓練資料正確率
print("測試資料正確率:",clf.score(test_data, test_label)) # 測試資料正確率

predict=clf.predict(test_data) # 預測
print("標準答案:",test_label) # 標準答案
print("預測答案:",predict) # 預測答案
```

```python
import graphviz  # 匯入 graphviz 套件（用來畫決策樹圖）

# 將訓練好的決策樹模型轉成 graphviz 格式資料
dot_data = tree.export_graphviz(clf, out_file=None)

# 建立圖形物件
graph = graphviz.Source(dot_data)

# 將圖輸出成檔案（預設為 PDF）
graph.render("iris")
# 會在目前資料夾產生 iris.pdf（決策樹圖）
```

```python
import pydotplus  # 匯入 pydotplus（用來處理 graphviz 圖形資料）
from IPython.display import Image  # 匯入 Image（在 Jupyter Notebook 顯示圖片）

# 將決策樹模型轉換成 dot 格式（graphviz 語言）
dot_data = tree.export_graphviz(
    clf,
    out_file=None,
    feature_names=iris.feature_names,  # 顯示每個特徵名稱（例如花瓣長度）
    class_names=iris.target_names,     # 顯示分類名稱（setosa, versicolor, virginica）
    filled=True,                       # 節點上色（不同分類顏色不同）
    rounded=True,                      # 節點邊角變圓
    special_characters=True            # 支援特殊字元
)

# 將 dot 資料轉成圖形物件
graph = pydotplus.graph_from_dot_data(dot_data)

# 產生 PNG 圖片並顯示（通常在 Jupyter Notebook 中使用）
Image(graph.create_png())
# 會直接在畫面中顯示決策樹圖
```
